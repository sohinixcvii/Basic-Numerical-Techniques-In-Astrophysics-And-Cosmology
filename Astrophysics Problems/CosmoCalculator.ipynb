```python
#importing necessary packages
from astropy.cosmology import FlatLambdaCDM as cosmo
import astropy.units as u
import numpy as np
import matplotlib.pyplot as plt
```

```python
#asking the user for important constants and converting them to float
h0=input("Enter the Hubble Constant in km/s/Mpc: ")
lam=input("Enter the cosmological constant: ")
om=input("Enter Omega matter: ")
z=input("Enter the redshift: ")
h0=float(h0)
lam=float(lam)
om=float(om)
z=float(z)

#defining our cosmological model
cm=cosmo(H0=h0*u.km/u.s/u.Mpc,Om0=om,)
```

```python
#calculating and displaying all the relevant data
print("We are assuming an LCDM universe here!")

#current age of the universe (redshift 0)
page=cm.age(0)
print("Present age of the universe is: "+str(page))

#age at given redshift
a=cm.age(z)
print("Age of the universe at the given redshift is: "+str(a))

#lookback time or light travel time
lt=cm.lookback_time(z)
print("Lookback time or light travel time is: "+str(lt))

#comoving volume
comv=cm.comoving_volume(z)
print("Comoving Volume at given redshift is: "+str(comv))

#radial comoving distance
comd=cm.comoving_distance(z)
print("Comoving Distance at given redshift is: "+str(comd))

#luminosity distance
dl=cm.luminosity_distance(z)
print("Luminosity Distance is: "+str(dl))

#angular diameter distance
da=cm.angular_diameter_distance(z)
print("Angular diameter distance is: "+str(da))
```
