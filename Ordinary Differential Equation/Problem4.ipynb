```python
#importing necessary packages
import math as m
import numpy as np
import matplotlib.pyplot as plt  
```

```python
v=700
DT= (100-0)/10000
g = 9.8
b = 4*10**(-5)

#defining a function to calculate projectile angle
def pa(th):
  vx0= v*np.cos(th)
  vy0= v*np.sin(th)
  X=[]
  Y=[]
  Vx= [vx0]
  Vy= [vy0]
  F= [b*v**2]
  TH= [th]
    
  for i in range(0,10000):
    if sum(Y)<0:
      break
    else:
      vx= Vx[-1] - F[-1]*np.cos(TH[-1])*DT      
      vy= Vy[-1] - g*DT - F[-1]*np.sin(TH[-1])*DT
    
      f= b*(vx**2 +vy**2)
      th= m.atan(vy/vx)

      TH.append(th)
      F.append(f)
      Vx.append(vx)
      Vy.append(vy)
  
      x= (Vx[-2] + Vx[-1])*DT/2 
      y= (Vy[-2] + Vy[-1])*DT/2

      X.append(x)
      Y.append(y)

  n=0
  xl=[]
  for i in range(0,len(X)):
    n= X[i]+n
    xl.append(n)

  M=0
  yl=[]
  for i in range(0,len(Y)):
      M= Y[i]+M
      yl.append(M)
  return plt.plot(xl, yl,label=str(int(-th*180/np.pi))+" degree");

for th in np.arange(20*np.pi/180,70*np.pi/180, 5*np.pi/180):
  pa(th)
  
plt.title("Trajectory of Different Projectiles for Different Launch Angles with Air Resistance")
plt.xlabel("Height")
plt.ylabel("Range")
plt.legend(loc="upper right")
plt.show()

print("We can the see the Range is Maximum for Firing Angles 35-40 Degrees")
```

```python
v= 700
g= 9.8
y0=0
t0=0
tn=130
n=1000

def ey(th,t0,y0,tn,n):
  def fy( th,t, y ):
    return v*np.sin(th) - g*t;
  y=[]   
  x=[] 
  h = (tn-t0)/n
  r=0
  while r<1000:
    r=r+1
    xr = v*np.cos(th)*((tn-t0)/n)*r
    x.append(xr)

  k=0
  while k<n:
    k=k+1
    slope = fy(th,t0,y0)
    yn = y0 + h * slope
    y.append(yn)      
    y0 = yn
    t0 = t0+h 

  Y = [0 if i < 0 else i for i in y]
  s = Y.count(0)
  x = x[:len(x)-s]
  Y = Y[:len(Y)-s]
   
  return plt.plot(x,Y,label=str(th*180/np.pi)+" degree");
```

```python
th=np.pi/4

ey(th,t0,y0,tn,n)
pa(th)
plt.title("Trajectory of a Projectile for Launch Angles equal to 45 degrees with and without Air Resistance")
plt.xlabel("Height")
plt.ylabel("Range")
plt.legend()
plt.show()
```
