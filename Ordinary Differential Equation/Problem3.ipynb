```python
#importing necessary packages
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

#defining the constants
v= 700
g= 9.8
y0,t0=0,0
tn=130
n=1000

#defining an array of the theta values
th = np.arange(20*(np.pi/180), 65*(np.pi/180), 5*(np.pi/180))

def ey(th,t0,y0,tn,n):
  def fy( th,t, y ):
    return v*np.sin(th) - g*t;
  y=[]   
  x=[] 
  h = (tn-t0)/n
  r=0
  while r<1000:
    r=r+1
    xr = v*np.cos(th)*((tn-t0)/n)*r
    x.append(xr)

  k=0
  while k<n:
    k=k+1
    slope =fy(th,t0, y0)
    yn = y0 + h * slope
    y.append(yn)      
    y0 = yn
    t0 = t0+h 

  Y = [0 if i < 0 else i for i in y]
  s = Y.count(0)
  x = x[:len(x)-s]
  Y = Y[:len(Y)-s]
  plt.plot(x,Y,label=str(int(th*180/np.pi))+" degree") 
  return;

  
for i in th:
  ey(i,t0,y0,tn,n)
plt.title("Trajectory of Different Projectiles for Different Launch Angles")
plt.xlabel("X-Distance")
plt.ylabel("Height")
plt.legend()
plt.show()
ran=[]
for i in th:
  r= (v**2)*np.sin(2*i)/g
  ran.append(r)

plt.plot(th*(180/np.pi), ran)

plt.title("Variation of Range with respect to Launch Angle")
plt.xlabel("Theta")
plt.ylabel("Range")
mx=max(ran)
ID=np.where(ran==mx)
yx=np.linspace(0.0,mx,100)
xx=[th[ID]*(180/np.pi)]*100
plt.plot(xx,yx,ls='--',c='black',label="maximum range")
plt.legend()
plt.show()
```

We can see here that for launch angle 45 degree, the range is maximum.
