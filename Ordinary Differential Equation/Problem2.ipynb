```python
#Trajectory of an elctron in a uniform magnetic field along z-axis.

from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
fig = plt.figure()  

q= 1.60217662*10**(-19)
m= 9.10938356*10**(-31)
c= 3*10**8

B= 10**(-4)

w=(q*B)/m
R=(m*2**(1/2))/(q*B)

t= np.linspace(0,10**(-6), 100)

z = 1*t
x = R*np.cos(w * t + 3*np.pi/4) + R/2**(1/2)
y = R*np.sin(w * t+ 3*np.pi/4) - R/2**(1/2)


ax = plt.axes(projection ='3d')  
ax.plot3D(x, y, z,c="red",marker='.')
ax.set_title("Trajectory of the Electron in a Magnetic Field along z-axis")
plt.show()
```

```python
#Euler method of integration

E= 5*10**(-3)
p= m*1

e_0 = ( (m**2)*(c**4) + 3*(p**2)*c**2 )**(1/2)
r= q*E*c

x0,y0,z0,t0=0,0,0,0

tn=10**(-5)
n=100

x= []
y= []
z= []
t= []

def f_x( t, x ):
    return (r*c*t + p*c**2)/(e_0**2 + (r*t)**2)**(1/2);

def f_y( t, y ):
    return (p*c**2)/(e_0**2 + (r*t)**2)**(1/2);

def f_z( t, z ):
    return (p*c**2)/(e_0**2 + (r*t)**2)**(1/2);
     
def ex(t0,x0,tn,n):    
  h = (tn-t0)/n
  for i in range(n):
      slope = f_x(t0, x0)
      xn = x0 + h * slope
      x.append(xn)
      t.append(t0)
      x0 = xn
      t0 = t0+h    
  x.append(xn)
  t.append(tn)
  return;

def ey(t0,y0,tn,n):    
  h = (tn-t0)/n
  for i in range(n):
      slope = f_y(t0, y0)
      yn = y0 + h * slope
      y.append(yn)
      t.append(t0)
      y0 = yn
      t0 = t0+h    
  y.append(yn)
 
  return;

def ez(t0,z0,tn,n):    
  h = (tn-t0)/n
  for i in range(n):
      slope = f_z(t0, z0)
      zn = z0 + h * slope
      z.append(zn)
      t.append(t0)
      z0 = zn
      t0 = t0+h    
  z.append(zn)
  
  return;

ex(t0,x0,tn,n)
ey(t0,y0,tn,n)
ez(t0,z0,tn,n)

ax = plt.axes(projection ='3d')  
ax.plot3D(x, y, z,c="red",marker='.')
ax.set_title("(Euler Method) Trajectory of the Electron in an Electric Field (Relativistic Motion)")
plt.show()
```
