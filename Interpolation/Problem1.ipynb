```python
#importing necessary packages
import numpy as np
import matplotlib.pyplot as plt

#defining the given data points
x=[1.0,2.1,5.0]
y=[8.0,20.6,13.7]

#Plotting the given points to visualize the points
plt.plot(x,y)

#fitting a first degree polynomial
c1=np.polyfit(x,y,1)
#fitting a second degree polynomial
c2=np.polyfit(x,y,2)
```

```python
#functions to find the y values for the polynomial fits
#to find y values for line fit
def line(x):
    y=c1[0]*x+c1[1]
    return y

#to find y values for the second degree fit
def par(x):
    y=c2[0]*x**2+c2[1]*x+c2[2]
    return y

#plotting the two fits and comparing
xt=np.linspace(1.0,5.0,10)
yl=line(xt)
yt=par(xt)
plt.plot(xt,yl,label="First degree polynomial fit",c="green")
plt.plot(xt,yt,label="Second degree polynomial fit")
plt.scatter(x,y,c="red",label="Given points",marker='x')
plt.title("Polynomial fit and given data comparison")
plt.legend()
```

## Using the line fit

```python
#asking the user to input a value of x
ip=input("Enter a value of x: ")

#Checking if the value of x is a valid value else displaying an error message
try:
    ip=float(ip)
except:
    print("Invalid Value. Try again.")
    quit()

#checking if given value of x lies within the pre specified range and proceeding to find the y value accordingly 
if ip<5.0 and ip>1.0:
    res=line(ip)
    print("Value of y at given x is: "+str(res))
    plt.plot(xt,yl,label="polynomial fit")
    plt.plot(ip,res,label="location of given point",marker='x')
    plt.title("Location of entered point and corresponding value")
    plt.ylabel("y values")
    plt.xlabel("x values")
    plt.legend()
else:
    print("Invalid value. Out of range")
```
