```python
#For limits -1 to 1, with an aplitude of 1 unit.
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad
import math as m

def f(x):
    return (2**0.5)/(2-2*x**2)**(1/2)
print("Limit: -1.0 to 1.0")
ans, err = quad(f, -1,1)
print("Answer is: "+str(ans)+" and error is: "+str(err))
```

```python
#For limits -0.5 to 0.5, with an aplitude of 0.5 unit.
def f2(x):
    return (2**0.5)/(0.5-2*x**2)**(1/2)
print("Limit: -0.5 to 0.5")
ans, err = quad(f2, -0.5,0.5)
print("Answer is: "+str(ans)+" and error is: "+str(err))
```

We can see above that even after changing the limits for from (-1,1) to (-0.5, 0.5) there is no change in amplitude, hence there is no dependency of amplitude on time period as it remains the same as pi.

```python
#For small values of amplitude, e^x becomes 1+x ignoring the later higher powers

def f3(x):
    return (2**0.5)/(2*(0.01)**2 - 2*x**2)**(1/2)

ans, err = quad(f3, -0.01,0.01)
print("For small values of amplitudes, answer is: "+str(ans)+" and error is: "+str(err))

#For diffent value of amplitude (small)

def f4(x):
    return (2**0.5)/(2*(0.02)**2 - 2*x**2)**(1/2)

ans, err = quad(f4, -0.02,0.02)
print("For a different (also small) value of amplitude, the answer is: "+str(ans)+" and error is: "+str(err))
```

Here we can see that for diferent values of amplitude, the time period comes the exaclty the same as that in the case of a SHM. 

```python
#For large values of amplitude

def f5(x):
    return (2**0.5)/(m.exp(2*(p/100)**2)-1 - (m.exp(2*x**2)-1))**(1/2)

ar=[]
for p in range(1,101):
    ans, err = quad(f5, -p/100,p/100)
    ar.append(ans)

x = np.linspace(0,1, 100)
y = ar
plt.title('Variation of Time Period w.r.t. Amplitude for Large Amplitudes')
plt.xlabel("Amplitude (meters)")
plt.ylabel("Time Period (seconds)")
plt.plot(x, y)
plt.show()
```
