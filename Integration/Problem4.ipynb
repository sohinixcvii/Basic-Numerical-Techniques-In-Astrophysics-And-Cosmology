```python
#imporitng the necessary packages
import math as m
import numpy as np

#2-point Gaussian quadrature method:

def f(x):
  return (x**2)*m.exp(-x**2);

def intg():

  y= f((1/3**(1/2))) + f(-(1/3**(1/2)))
  print("Result using 2 point Gaussian Quadrature is: "+str(y))

  return;

intg()
```
