```python
#importing necessary packages
import numpy as np
import matplotlib.pyplot as plt
import sympy as sy
import math as m
```

```python
#defining the four different integrands sequentially
def f1(x):
    return x

def f2(x):
    return x**2

def f3(x):
    return [m.sin(xi) for xi in x]

def f4(x):
    return [xi*m.sin(xi) for xi in x]
```

```python
'''defining the funtion for trapezoidal rule
x: the values over which the integrand needs to be integrated.
c: an integer to indicate which integrand needs to be integrated.
'''
def trap(x,c):
    h=(max(x)-min(x))/len(x)
    n=len(x)
    X=sy.Symbol('X')
    if c==1:
        y=f1(x)
        d=0
    if c==2:
        y=f2(x)
        d=2
    if c==3:
        y=f3(x)
        d=-m.sin(np.mean(x))
    if c==4:
        y=f4(x)
        d=-np.mean(x)*m.sin(np.mean(x))+2*m.cos(np.mean(x))
    s=0
    for i in range(1,n-1):
        s=s+y[i]
    I=(h/2)*(y[0]+2*s+y[-1])
    E=-((max(x)-min(x))/12)*h**2*d
    return [I,E]
```

```python
'''Defining a function for the Simpson's 1/3rd rule (commonly known as Simpson's rule)
x: the values over which the integrand needs to be integrated.
c: an integer to indicate which integrand needs to be integrated.
'''
def simp(x,c):
    h=(max(x)-min(x))/len(x)
    n=len(x)
    X=sy.Symbol('X')
    if c==1:
        y=f1(x)
        d=0
    if c==2:
        y=f2(x)
        d=0
    if c==3:
        y=f3(x)
        d=m.sin(np.mean(x))
    if c==4:
        y=f4(x)
        d=np.mean(x)*m.sin(np.mean(x))-4*m.cos(np.mean(x))
    s1=0
    s2=0
    for i in range(1,n-1):
        if i%2==0:
            s2=s2+y[i]
        else:
            s1=s1+y[i]
    I=(h/3)*(y[0]+4*s1+2*s2+y[-1])
    E=(max(x)-min(x))*180*(h**4)*d
    return [I,E]
```

```python
#creating an array to store the N=2,4,8...1024 values
i=2
ar=[]
while i<=1024:
    ar.append(i)
    i=i*2
er=[]

#Integrating the problems one by one
print("Integrating first problem using Trapezoidal rule. \n")
for el in ar:
    x=np.linspace(0.0,1.0,el)
    res=trap(x,1)
    er.append(res[1])
    print("Integration result with "+str(el)+" intervals is: "+str(res[0])+" and error"+str(res[1]))
plt.plot(ar,er)
plt.title("Error plot")
plt.xlabel("N values")
plt.ylabel("Error values")
plt.show()
```
