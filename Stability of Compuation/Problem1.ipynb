```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#single precision
print("Single precision")
s=1.0
x=np.linspace(1.0,100.0,100)
for k in x:
    s=0.5*s
    t=s+1.0
    if t<=1.0:
        s=2.0*s
        print(str(k-1)," ",str(s))
```

```python
#double precision
print("Double Precision")
s=1.00
x=np.linspace(1.00,100.00,100)
for k in x:
    s=0.50*s
    t=s+1.00
    if t<=1.00:
        s=2.00*s
        print(str(k-1)," ",str(s))
```
