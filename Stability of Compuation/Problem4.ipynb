```python
import numpy as np
import sympy as sy
import math

#Function to integrate using inbuilt integration function
def intg(x,n):
    y=(x**n)*(sy.exp(x))
    return y

#Function to find yn using recurrence relation
def rec(n):
    x=sy.Symbol('x')
    yn=sy.integrate(intg(x,n-1),(x,0,1))
    y=math.exp(1)-(n)*yn
    return y
```

```python
x=sy.Symbol('x')
i15=rec(15)
i20=rec(20)
res15=sy.integrate(intg(x,15),(x,0,1))
res20=sy.integrate(intg(x,20),(x,0,1))
i15=float(i15)
i20=float(i20)
res15=float(res15)
res20=float(res20)
e15=(res15-i15)/res15
e20=(res20-i20)/res20
```

```python
print("Integrated value using recurrence relation for n=15 is: ")
print(i15)
print("\n Integrated value without using recurrence relation for n=15 is: ")
print(res15)
print("Error for using recurrence relation for n=15 is: "+str(e15*100)+"%")
print("Integrated value using recurrence relation for n=20 is: ")
print(i20)
print("\n Integrated value without using recurrence relation for n=20 is: ")
print(res20)
print("Error for using recurrence relation for n=20 is: "+str(e20*100)+"%")
```

As is evident from the above analysis, huge errors are incurred by using the recurrence formula for integration. Hence, the recurrence formula is unstable.
