```python
import numpy as np

x=(np.sqrt(5.0)-1.0)/2.0

def rec(n):
    y=x**(n-1)-x**n
    return y
res=rec(16)
er=(res-x**16)/x**16
print("Value of x is: "+str(x))
print('x^16'," by recurrence relation is: "+str(res)+"\n Value without recurrence relation is: "+str(x**16)+"\n Error: "+str(er*100)+"%")

res=rec(30)
er=(res-x**30)/x**30
print("x^30"+" by recurrence relation is: "+str(res)+"\n Value without recurrence relation is: "+str(x**30)+"\n Error: "+str(er*100)+"%")

res=rec(40)
er=(res-x**40)/x**40
print("x^40"+" by recurrence relation is: "+str(res)+"\n Value without recurrence relation is: "+str(x**40)+"\n Error: "+str(er*100)+"%")

res=rec(50)
er=(res-x**50)/x**50
print("x^50"+" by recurrence relation is: "+str(res)+"\n Value without recurrence relation is: "+str(x**50)+"\n Error: "+str(er*100)+"%")
```

Hence it is clear that there is a consistent error of about 38% magnitude in the recurrence relation. It systematically gives a smaller value than the original value. Hence, this relation is unstable.
