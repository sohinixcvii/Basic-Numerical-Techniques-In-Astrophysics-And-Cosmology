```python
#importing necessary packages
import matplotlib.pyplot as plt
import numpy as np
import random as rn

#defining a function for drunken walk
def dwalk (st,re):
  xs=[]
  p=[]
  r=[-1,1]
  for j in range(0,re):
    p.clear()
    for i in range(0,st):
      n=rn.choice(r)
      p.append(n)
    xs.append(sum(p))
    xsp = [0 if i < 0 else i for i in xs]  
    xsn = [0 if i >= 0 else i for i in xs]

    x = range(100)

  fig = plt.figure()
  ax = plt.subplot(111)
  plt.figure(figsize=(200,10 ))
  ax.bar(x, xsn, width=1, color='r')
  ax.bar(x, xsp, width=1, color='b')
  return ;

dwalk (3600, 100)


```
