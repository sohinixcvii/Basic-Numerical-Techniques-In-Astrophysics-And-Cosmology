```python
#importing necessary packages
import matplotlib.pyplot as plt
import numpy as np
import random as rn

#defining a function for drunken walk
def dwalk(st):
  p=[]
  r=[-1,1]
  for i in range(0,st):
    n=rn.choice(r)
    p.append(n)  
    s=sum(p)
  mn=s/st   
  return mn;

mn=[]
for i in range(1,1000):
  mn.append(dwalk(i))

plt.figure(figsize=(20, 5))
plt.plot(range(1,1000),mn,c="black")
plt.xlabel('Time(steps)')
plt.ylabel('Mean')
plt.title('Mean Vs Steps')  

plt.show()

print("The mean tends to 0 with increasing number of steps whereas the RMS is always 1")
```
