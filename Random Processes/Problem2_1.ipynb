```python
#importing the necessary packages
import numpy as np
import statistics as st
import matplotlib.pyplot as plt

std=[]
intg=[]

for i in range(2,1000):

  a = 0
  b = np.pi
  x = np.random.uniform(a,b,i)
  fx = np.cos(x)

  std.append(st.stdev(fx))
  intg.append(np.mean(fx)*(b-a))

plt.figure(figsize=(20, 5))
plt.title('f(x)=cos(x)')
plt.plot(range(2,1000),intg, label='Integration Value')
plt.plot(range(2,1000), std, label='Standard Deviation')
plt.legend(framealpha=1, frameon=True);
plt.xlabel('Number of Random Points (N)')
plt.ylabel('Value')


plt.show()
```

```python
std1=[]
intg1=[]

for i in range(2,1000):

  a = 0
  b = 1
  x = np.random.uniform(a,b,i)
  fx = x**2

  std1.append(st.stdev(fx))
  intg1.append(np.mean(fx)*(b-a))

plt.figure(figsize=(20, 5))
plt.title('$f(x)=x^2$')
plt.plot(range(2,1000), intg1, label='Integration Value')
plt.plot(range(2,1000), std1, label='Standard Deviation')
plt.legend(framealpha=1, frameon=True);
plt.xlabel('Number of Random Points (N)')
plt.ylabel('Value')


plt.show()
```
