```python
import numpy as np
import statistics as st
import matplotlib.pyplot as plt

std=[]
intg=[]

for i in range(0,1000):

  a = 0
  b = 1
  x = np.random.uniform(a,b,500)
  fx = x**2

  std.append(st.stdev(fx))
  intg.append(np.mean(fx)*(b-a))

print("Sigma divided by square root of N is:",st.mean(std)/500**(1/2) )
print("Standard Deviation of Integration is (Sigma M)",st.stdev(intg))
print("Evidently, both the values are almost the same")
```
